# Handling the NaN files

```python
df1.fillna(value=0, inplace=True)
```

# Preprocessing the Age Column

```python
def updateAgeTable():
    for i in range(len(df1)):
        ages = df1["age"].iloc[i].split("-")
        
        if len(ages) > 1:
            actualAge = (int(ages[1]) + int(ages[0]))/2
        else:
            actualAge = float(ages[0])
        
        df1["age"].iloc[i] = ceil(actualAge)
```

```python
updateAgeTable()
```

# Pre Processing the Gender Column

```python
df1["sex"] = df1["sex"].replace(to_replace=["male","female"],value=[0,1])
```

# Pre Processing all the Symptom columns

```python
columns = df1.columns
columns
```

```python
def updateSymptomsTable():
    for i in range(3,len(columns)-1):
        for j in range(len(df1)):
            if df1[columns[i]].iloc[j] != 0:
                df1[columns[i]].iloc[j] = 1
            else:
                continue
    
```

```python
df1.head()
```

```python
updateSymptomsTable()
```

```python
df1.head()
```

# Pre Processing the Output Label

```python
x = df1["covid_19"].unique()
```

```python
x
```

```python
type(x[0])
```

```python
df1["covid_19"] = df1["covid_19"].replace(to_replace=[x[0],x[1]],value=[1,0])
```

# Shuffling all the rows

```python
df1 = df1.sample(frac=1).reset_index(drop=True)
```

```python
df1.drop("ID",inplace=True, axis=1)
```
